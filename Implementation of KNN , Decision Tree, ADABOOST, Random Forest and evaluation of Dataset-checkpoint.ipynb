# IMPLEMENTING MACHINE LEARNING ALGORITHMS ON WINE DATASET FROM UCI : https://archive.ics.uci.edu/ml/datasets/Wine

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt
#read file and plot a histogram for each class
open('wineData.csv')
dt = pd.read_csv('wineData.csv')
dt['Class'].value_counts().plot(kind = 'bar',color = 'turquoise')
print(dt.describe())
#plt.savefig('Bar_PLOT.png')
import seaborn as sns
sns.set(style='ticks')
sns.pairplot(dt,hue='Class')
#Map the classes to 0 and 1
New_mapping = {label : idx for idx,label in enumerate(np.unique(dt["Class"]))}
dt['Class'] = dt['Class'].map(New_mapping)
#Normalize to [0,3] scale
dtTemp = dt.iloc[:,1:]
x = dt.iloc[:,1:].values
min_max_scaler = preprocessing.MinMaxScaler(feature_range = (0,3))
x_scaled = min_max_scaler.fit_transform(x)
NormalizedDT = pd.DataFrame(x_scaled,columns=dtTemp.columns)
NormalizedDT['Class'] = dt['Class']
export_csv = NormalizedDT.to_csv (r'wineNormalized.csv', index = None, header=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
dt=pd.read_csv('wineNormalized.csv')

labels = []
for name,series in dt.loc[:,'Alcohol':'Proline'].iteritems():
    k=name
    labels.append(k)
#break into training and testing sets
X,y = dt.iloc[:,:13].values, dt['Class'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state = 0,stratify = y)
df1 = pd.DataFrame(data= X_train)
df2 = pd.DataFrame(data= y_train,columns=['Class'])
df3 = pd.DataFrame(data= X_test)
df4 = pd.DataFrame(data= y_test,columns=['Class'])
df1.columns=[labels]
df3.columns=[labels]
training = df1
training['Class'] = df2
testing = df3
testing['Class'] = df4
export_csv = training.to_csv (r'Training_set_stratified.csv', index = None, header=True)
export_csv = testing.to_csv (r'Testing_set_stratified.csv', index = None, header=True)
#implement winning decision tree
from sklearn.tree import DecisionTreeClassifier
resultsEntropy=pd.DataFrame(columns=['LevelLimit' , 'Score for Training' , 'Score for Testing'])
#from sklearn.tree import export_graphviz
for treeDepth in range(1,6):
    dct=DecisionTreeClassifier(criterion='gini',max_depth=treeDepth,random_state=0)
    dct=dct.fit(X_train, y_train)
#    export_graphviz(dct,out_file='tree'+str(treeDepth)+'.dot')
    dct.predict(X_test)
    scoreTrain=dct.score(X_train, y_train)
    scoreTest=dct.score(X_test, y_test)
    resultsEntropy.loc[treeDepth]=[treeDepth, (scoreTrain), (scoreTest)]

print(resultsEntropy)
resultsEntropy.pop('LevelLimit')
ax=resultsEntropy.plot()
plt.savefig('Decision_tree.png')
```

```python
import pandas as pd
#import matplotlib.pyplot as plt
dt = pd.read_csv('wineNormalized.csv')
training = pd.read_csv('Training_set_stratified.csv')
testing = pd.read_csv('Testing_set_stratified.csv')
X_train = training.iloc[:,:13]
X_test = testing.iloc[:,:13]
y_train = training.iloc[:,13]
y_test = testing.iloc[:,13]

#KNN
from sklearn.neighbors import (KNeighborsClassifier,NeighborhoodComponentsAnalysis)
from sklearn.pipeline import Pipeline
resultsknn = pd.DataFrame(columns=['KNN','Score for Training','Score for Testing'])
resultsnca = pd.DataFrame(columns=['NCA','Score for Training','Score for Testing'])
for KnnCount in range(1,20):
    knn = KNeighborsClassifier(n_neighbors = KnnCount, p=2, metric = 'minkowski', weights = 'distance')
    nca = NeighborhoodComponentsAnalysis(random_state=42)
    knn.fit(X_train,y_train)
    nca_pipe = Pipeline([('nca', nca), ('knn', knn)])
    nca_pipe.fit(X_train, y_train) 
    scoreTrain = knn.score(X_train,y_train)
    scoreTest =knn.score(X_test,y_test)
    nca_score = nca_pipe.score(X_test, y_test)
    nca_score_train = nca_pipe.score(X_train, y_train)
    resultsknn.loc[KnnCount] = [KnnCount,scoreTrain,scoreTest]
    resultsnca.loc[KnnCount] = [KnnCount,nca_score_train,nca_score]
print(resultsknn.head(11))
print(resultsnca.head(11))
resultsknn.pop('KNN')
ax = resultsknn.plot()
resultsnca.pop('NCA')
ax = resultsnca.plot()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
dt = pd.read_csv('wineNormalized.csv')
ndt = pd.DataFrame(columns=['Proline','Alcohol','Color intensity','Flavanoids'])
#make new derived feaatures
ndt['Proline']= dt['Proline']
ndt ['Alcohol']=dt['Alcohol']
ndt ['Color intensity']=dt['Color intensity']
ndt ['Flavanoids']=dt['Flavanoids']
ndt['Alco/pro'] = ndt['Alcohol']/ndt['Proline']
ndt['col/pro'] = ndt['Color intensity']/ndt['Proline']
ndt['flav/pro'] = ndt['Flavanoids']/ndt['Proline']
ndt['Flav/Col'] = ndt['Flavanoids']/ndt['Color intensity']
ndt['Flav/álco'] = ndt['Flavanoids']/ndt['Alcohol']
ndt['Col/alco'] = ndt['Color intensity']/ndt['Alcohol']
ndt['Col+Alc'] = ndt['Color intensity']/ndt['Alcohol']
ndt['Alc+Flav'] = ndt['Alcohol']+ndt['Flavanoids']
ndt['Alc+Pro'] = ndt['Alcohol']+ndt['Proline']
ndt ['Class'] = dt['Class']
#clamp them for any outliers
df_clamp=pd.DataFrame()
df_repo=ndt.describe()
df_inv=df_repo.transpose()
df_clamp['Lower']=df_inv['25%']-df_inv['25%'] 
df_clamp['Upper']=(df_inv['75%']/df_inv['75%'])*3
for name,series in ndt.iteritems():
    for i in ndt[name].index.values:
        if ndt.loc[i,name] < df_clamp.loc[name,'Lower']:
            ndt.loc[i,name]=ndt.loc[name,'Lower']
        elif ndt.loc[i,name]>df_clamp.loc[name,'Upper']:
            ndt.loc[i,name]=df_clamp.loc[name,'Upper']
        else:
            ndt.loc[i,name]=ndt.loc[i,name]
#stratify and make training and testing sets
#ndt.pop('Alc+Pro')
#ndt.pop('Alc+Flav')
X,y = ndt.iloc[:,:13].values, ndt['Class'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state = 0,stratify = y)
df1 = pd.DataFrame(data= X_train)
df2 = pd.DataFrame(data= y_train,columns=['Class'])
df3 = pd.DataFrame(data= X_test)
df4 = pd.DataFrame(data= y_test,columns=['Class'])
print(ndt.columns.to_list())
labels = []
for name,series in ndt.loc[:,'Proline':'Alc+Pro'].iteritems():
    k=name
    labels.append(k)
df1.columns=[labels]
df3.columns=[labels]
training = df1
training['Class'] = df2
testing = df3
testing['Class'] = df4
X_train = training.iloc[:,:13]
X_test = testing.iloc[:,:13]
y_train = training.iloc[:,13]
y_test = testing.iloc[:,13]
#feature importance
from sklearn.ensemble import RandomForestClassifier
feat_labels =ndt.columns[:13]
forest = RandomForestClassifier(criterion='gini',n_estimators = 10000, random_state = 0, n_jobs=-1)
forest.fit(X_train,y_train)
importances = forest.feature_importances_
indices = np.argsort(importances)[::-1]
results=pd.DataFrame(columns=['Count of Trees','Score for training','Score for testing'])
F_I=pd.DataFrame(columns=['Feature','Importance'])
for f in range(X_train.shape[1]):
    print("%d. %.*s %f"%(f+1,30,feat_labels[indices[f]],importances[indices[f]]))
    F_I=F_I.append({'Feature':feat_labels[indices[f]],'Importance':importances[indices[f]]},ignore_index = True)
#F_I.to_csv('feature_modified.csv')
#c.to_csv('Derived_features.csv')
#remove bad features
X_train.pop('Col+Alc')
X_test.pop('Col+Alc')
X_train.pop('Flav/Col')
X_test.pop('Flav/Col')
X_train.pop('Flav/álco')
X_test.pop('Flav/álco')
X_train.pop('Alc+Flav')
X_test.pop('Alc+Flav')
X_train.pop('flav/pro')
X_test.pop('flav/pro')
X_train.pop('Alco/pro')
X_test.pop('Alco/pro')
X_train.pop('col/pro')
X_test.pop('col/pro')
X_train.pop('Col/alco')
X_test.pop('Col/alco')
#implement randome forest
indexR=1
for size in range(1,15,2):
    forest=RandomForestClassifier(criterion='gini',n_estimators=size,max_depth=3,random_state=0)
    forest.fit(X_train,y_train)
    scoreTrain=forest.score(X_train,y_train)
    scoreTest=forest.score(X_test,y_test)
    results.loc[indexR]=[size,scoreTrain,scoreTest]
    indexR=indexR+1
print(results)
results.pop('Count of Trees')
ax=results.plot()
import seaborn as sns
sns.set(style='ticks')
sns.pairplot(ndt,hue='Class')
#plt.savefig('rf_modified.png')
```
